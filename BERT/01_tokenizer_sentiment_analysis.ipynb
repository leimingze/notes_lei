# tokenizer,构造输入

- tokenizer,model:相匹配,tokenzier outputs=>model input
- tokenizer
    - len(input_ids)==len(attention_mask)
    - tokenizer.tokenize(text) 将字符串切分为 token 列表
    - tokenizer.convert_tokens_to_ids(tokens) 将 token 列表转换为 对应的 ID 列表
    - tokenizer.encode(text) 一步完成 tokenize + convert_to_ids，返回 token ID 序列
    - tokenizer(text) 完整的 tokenizer 调用，返回带有 input_ids, attention_mask 等结构的字典

```python
test_senteces=['today is not that bad','today is so bad']
model_name= 'distilbert-base-uncased-finetuned-sst-2-english'
```

```python
from transformers import AutoTokenizer,AutoModelForSequenceClassification
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model=AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
batch_input=tokenizer(test_senteces,max_length=256,truncation=True,padding='max_length',return_tensors='pt')
#truncation超过最大长度截取，padding补0，return_tensors 返回pytorch张量
```

```python
import torch
import torch.nn.functional as F
```

```python
model.config
```

# model，调用模型

```python
with torch.no_grad():
    outputs=model(**batch_input)
    print(outputs)
    scores=F.softmax(outputs.logits,dim=1)
    print(scores)
    label=torch.argmax(scores,dim=1)# 取概率最高的类别作为预测结果
    print(label)
    res=[model.config.id2label.get(id) for id in label.tolist()]
    print(res)
```
